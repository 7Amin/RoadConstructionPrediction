```python
import pandas as pd
import requests 
import os
from xml.etree import ElementTree
from datetime import datetime
```

```python
chunk_size = 12 ** 6
TrafficEvents = None

TrafficEvents = {
    "name": "TrafficEvents",
    # "file_name": "TrafficEvents_Aug16_Dec19_Publish_Road_Construction.csv",
    "file_name": "All_Constructions_June2020_Combined_TO_StartEnd_30_100.csv",
    "path": "/home/amin/CETI/RoadConstruction/TrafficEventData/",
    "part_path": "part"
}

number_of_records_in_each_file = 500
number_of_folders = 10
number_of_file_in_folders = 411
```

```python
def read_big_data_by_filter_with_key_values(path, filters):
    data = None
    print("method {} is started ".format("read_big_data_filter_with_key_values"))
    print("read csv data is started with file {}".format(path))
    print("chunk size is {}".format(chunk_size))
    for key, value in zip(filters['keys'], filters['values']):
        print("the key is {} and the value is {}".format(filters['keys'], filters['values']))
    for num, df in enumerate(pd.read_csv(path, chunksize=chunk_size), start=1):
        print("continue reading file page num is {}".format(num))
        for key, value in zip(filters['keys'], filters['values']):
            df = df[df[key] == value]
        data = df.append(data)
    print("method {} finished ".format("read_big_data_filter_with_key_values"))
    return data
```

```python
traffic_file_path = TrafficEvents["path"] + TrafficEvents["file_name"]

traffic_filters = {
    "keys": ["Type"],
    "values": ["Construction"],
    }

traffic_events_data = read_big_data_by_filter_with_key_values(traffic_file_path, traffic_filters)
```

```python
print("shape of data is {}".format(traffic_events_data.shape))
print(number_of_file_in_folders)
columns = []
for col in traffic_events_data.columns: 
    columns.append(col)

for folder_index in range(number_of_folders):
    dir = os.path.join(TrafficEvents["path"], TrafficEvents["part_path"], str(folder_index))
    print(dir)
    print("folder_index: {}".format(folder_index))
    if not os.path.exists(dir):
        os.mkdir(dir)

    for file_index in range(number_of_file_in_folders):
        print("file_index: {}".format(file_index))
        base_index = (folder_index *  number_of_records_in_each_file * number_of_file_in_folders)
        start_point_record = base_index + file_index * number_of_records_in_each_file
        end_point_record = base_index + (file_index + 1) * number_of_records_in_each_file
        
        print("start is {} and end is {}".format(start_point_record, end_point_record))
        df = pd.DataFrame(columns=columns)
        for index in range(start_point_record, end_point_record): 
            data = traffic_events_data.iloc[index]
            df = df.append(data, ignore_index=False)
        print(df.shape)
        file_url = dir + "/file_{}.csv".format(file_index)
        df.to_csv (file_url, index = False, header=True)
        print(len(df))
        
print("end")
```
